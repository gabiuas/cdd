```python
dados13 = pd.read_csv('DOM2013csv.zip', compression='zip', header=0, sep=',', quotechar='"')
dados14 = pd.read_csv('DOM2014csv.zip', compression='zip', header=0, sep=',', quotechar='"')
dados15 = pd.read_csv('DOM2015csv.zip', compression='zip', header=0, sep=',', quotechar='"')
```

```python
df13= pd.DataFrame({"UF":dados13.UF})


#COLUNAS
df13['RU']=dados13.V4105
df13['Tv_cor']=dados13.V0226
df13['Tv_fina']=dados13.V02270
df13['Tv_tubo']=dados13.V02271
df13['tv_ass']=dados13.V02272
df13['dg_abert']=dados13.V02273
df13['tv_para']=dados13.V02274
df13['dvd']=dados13.V2027
df13['micro']=dados13.V0231
df13['tablet']=dados13.V02321
df13['internet']=dados13.V02322
df13['mes_domici']=dados13.V4614
df13['dom_per_capt']=dados13.V4622



#RESPOSTAS
df13.Tv_cor = df13.Tv_cor.astype('category')
df13.Tv_cor.cat.categories = ('Sim','Não')

df13.dvd = df13.dvd.astype('category')
df13.dvd.cat.categories = ('Sim','Não')

df13.tv_para = df13.tv_para.astype('category')
df13.tv_para.cat.categories = ('Sim','Não','NS')

df13.tv_ass = df13.tv_ass.astype('category')
df13.tv_ass.cat.categories = ('Sim','Não')

df13.dg_abert = df13.dg_abert.astype('category')
df13.dg_abert.cat.categories = ('Sim','Não','NS')

df13.micro = df13.micro.astype('category')
df13.micro.cat.categories = ('Sim','Não')

df13.tablet = df13.tablet.astype('category')
df13.tablet.cat.categories = ('Sim','Não')

df13.internet = df13.internet.astype('category')
df13.internet.cat.categories = ('Sim','Não')

df13.dom_per_capt = df13.dom_per_capt.astype('category')
df13.dom_per_capt.cat.categories = ('Zero','Zero - 1/4','1/4 - 1/2','1/2 - 1','1 - 2','2 - 3','3 - 5','5+','SD')


#EXIGÊNCIAS
df13[((df13.UF==31)|(df13.UF==32)|(df13.UF==33)|(df13.UF==35))&(df13.mes_domici>1000)].sort_values(by='mes_domici').head(10).dropna()
#dados.sort_values(by="EDUCACAO").head()

#df13.iloc[0:11,[0,1,2,5,6,7,8,9,10,11,13]]

df13L=df13.iloc[:,[0,1,2,5,6,7,8,9,10,11,13]]
df13T=df13.iloc[:,[12]]
df13Tv=df13.iloc[:,[3,4]]


df13[((df13.UF==31)|(df13.UF==32)|(df13.UF==33)|(df13.UF==35))&(df13.mes_domici>1000)].sort_values(by='mes_domici').head(10).dropna()

```
