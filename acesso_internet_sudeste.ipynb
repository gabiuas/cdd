## Introdução


   Este relatório teve como principal objetivo analisar o acesso a Internet. A amostra da análise foi composta apenas pela população do Sudeste do Brasil (SP, ES, MG e RJ). A pergunta que guiou a análise foi: Como o acesso da população do Sudeste, a internet, varia conforme sua renda? A partir deste objetivo, outros questionamentos começaram a aparecer e se tornar interessantes, tal como: Por quais aparelhos eletrônicos a população do Sudeste mais acessa a internet?
   
   Com o objetivo traçado e as análises concluídas, mostrou-se interessante analisar os dados de não apenas um ano, mas sim de três anos em sequência (2013, 2014 e 2015), para que comparações, semelhânças e diferenças pudesse ser detectadas, e melhorias apontadas, permitindo que perspectivas de melhoras e pioras pudessem ser concretizadas.

## Histogramas Salários

 É possivel notar que grande parte da amostra tem uma renda domiciliar per capta entre mil e 4 mil, porém existem algumas pessoas que ganham muito mais tornando os gráficos, dos três anos, assimétricos a direita o que resulta em uma distorção da média de salários nos três casos.
